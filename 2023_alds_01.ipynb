```python
import matplotlib.pyplot as plt  # „Ç∞„É©„Éï„ÇíÊèè„Åè„Åü„ÇÅ„ÅÆ„É¢„Ç∏„É•„Éº„É´
import math             # log„Çí‰Ωø„ÅÜ„Åü„ÇÅ„ÅÆ„É¢„Ç∏„É•„Éº„É´
```

```python
n_list = [10**x for x in range(1,6)]

ct_list =[]

############################
# lg(n)
ct_lgn  = []
for n in n_list:
  ct = math.log2(n) * (10**-6)
  ct_lgn.append(ct)
print(ct_lgn)
ct_list.append(ct_lgn)


############################
# n^(-0.5)
ct_sqn  = []
for n in n_list:
  ct_sqn.append((n**0.5)  * (10**-6))
print(ct_sqn)
ct_list.append(ct_sqn)


############################
# n
ct_n  = [n *(10**-6) for n in n_list]
print(ct_n)
ct_list.append(ct_n)

```

„ÄêÂïè4„ÄëÂÜçÂ∏∞Èñ¢Êï∞„ÇíÁî®„ÅÑ„Å¶Èöé‰πó„ÇíË®àÁÆó„Åô„ÇãÔºé

$n!$„ÅØÔºåÊ¨°„ÅÆ„Çà„ÅÜ„Å´ÂÆöÁæ©„Åß„Åç„ÇãÔºé
$$
\begin{cases}
  n!=n \cdot (n-1)! & n>0 \\
  0!=1 &
\end{cases}
$$

Èöé‰πó„ÅÆË®àÁÆó„Çí„Åô„Çã„Åü„ÇÅ„ÅÆËá™‰ΩúÈñ¢Êï∞Factorial„Çí‰ΩúÊàê„Åô„ÇãÔºé
- Êï¥Êï∞ùëõ„ÇíÂèó„ÅëÂèñ„ÇãÔºé
- ùëõ„Åå0„Å™„Çâ„Å∞Ôºå1„Çíreturn„Åô„ÇãÔºé
- ùëõ„Åå1‰ª•‰∏ä„Å™„Çâ„Å∞Èñ¢Êï∞Factorial„ÇíÁî®„ÅÑ„Å¶(ùëõ‚àí1)!„ÇíË®àÁÆó„ÅóÔºå„Åù„ÅÆËøî„ÇäÂÄ§„Å®ùëõ„ÇíÊéõ„Åë„Åü„ÇÇ„ÅÆ„Çíreturn„Åô„ÇãÔºé




```python
def Factorial(n):
  if n>0:
    return n*Factorial(n-1)
  return 1
```

„ÄêÂïè5„Äë2„Å§„ÅÆÊï¥Êï∞ùëö,ùëõ„ÅÆÊúÄÂ§ßÂÖ¨Á¥ÑÊï∞„ÇíÊ±Ç„ÇÅ„Çã„Åü„ÇÅ„ÅÆÊñπÊ≥ï„Å®„Åó„Å¶Ôºå„É¶„Éº„ÇØ„É™„ÉÉ„Éâ„ÅÆ‰∫íÈô§Ê≥ï„Åå„ÅÇ„ÇãÔºé

„É¶„Éº„ÇØ„É™„ÉÉ„Éâ„ÅÆ‰∫íÈô§Ê≥ï„ÅØ‰∏ã„ÅÆ„Çà„ÅÜ„Å´ÂÜçÂ∏∞„ÇíÁî®„ÅÑ„Å¶ÂÆöÁæ©„Åß„Åç„ÇãÔºé

$$
gcd(m,n) =
\begin{cases}
  gcd‚Å°(ùëö‚àíùëõ,ùëõ)  & \textrm{if}\  m>n\\
  gcd(ùëö,ùëõ‚àíùëö) &  \textrm{if}\  m<n\\
  ùëö &  \textrm{if}\  m=n
\end{cases}
$$

2„Å§„ÅÆÊï¥Êï∞„ÅÆÊúÄÂ§ßÂÖ¨Á¥ÑÊï∞„ÇíÂÜçÂ∏∞„ÇíÁî®„ÅÑ„Å¶Ê±Ç„ÇÅ„ÇãËá™‰ΩúÈñ¢Êï∞gcd„Çí‰ΩúÊàê„Åó„Å¶„Åè„Å†„Åï„ÅÑÔºé

```python
def gcd(m,n):
  if m>n:
    return gcd(m-n, n)
  elif m<n:
    return gcd(m,n-m)
  return m
```

```python
m,n = 24,18
print(str(m)+"„Å®"+str(n)+"„ÅÆÊúÄÂ§ßÂÖ¨Á¥ÑÊï∞„ÅØ"+str(gcd(24,18))+"„Åß„Åô")
```
