```python
import pulp
import random
```

まずは前回と同じようにPuLPライブラリをインストール．その後import．

復習として，第6回演習課題の問1を解いてみる．

問題（lp1）と，製品A~Cの生産量を表す変数X1〜X3を登録する．
生産量が負の値になることはないので，lowBoundは0，upBoundはその製品の需要量とする．

```python
lp1 = pulp.LpProblem('lp1', sense=pulp.LpMaximize)
X1 = pulp.LpVariable('X1', lowBound=0, upBound=2000)  # LowBoud=0を指定することで制約条件（４）を表現できる。
X2 = pulp.LpVariable('X2', lowBound=0, upBound=3000)
X3 = pulp.LpVariable('X3', lowBound=0, upBound=5000)
```

目的関数と，その他の制約式を追加．

```python
lp1+= 50 * X1 + 80 * X2 + 100 * X3

for C in [[4,4,10,60000],[5,12,6,60000]]:
  lp1 += C[0] * X1 + C[1] * X2 + C[2] * X3 <= C[3]


print(lp1)
```

実際に解いてみる．

```python
status = lp1.solve()
# print(pulp.LpStatus[status])

print('Optimal solution: (X1, X2, X3) = ({}, {}, {})'.format(X1.value(), X2.value(), X3.value()))
print('Objective function value: f = {}'.format(lp1.objective.value()))
```

```python
lp4 = pulp.LpProblem('lp4', sense=pulp.LpMaximize)

A,X,C,UL=[],[],[],[]

number_of_decision_variables = 500
number_of_constraints =100

for i in range(number_of_decision_variables):
  X.append(pulp.LpVariable('X{}'.format(i+1), lowBound=0, upBound=random.randint(10,30),cat="Integer"))
  A.append(random.randint(1,10))


for j in range(number_of_constraints): # 制約式の本数
  UL.append(random.randint(number_of_decision_variables, 2*number_of_decision_variables))
  c=[]
  for i in range(number_of_decision_variables):
    c.append(random.randint(1,3))
  C.append(c)

of=None # Objective Function
for i in range(len(X)):
  of+=A[i]*X[i]
lp4 += of

for j,c in enumerate(C):
  eq=None
  for i in range(len(X)):
    eq+=c[i]*X[i]
  lp4+= eq <= UL[j]

# print(lp4)
status = lp4.solve()

print('\n\nOptimal solution')
for i in range(len(X)):
  print("X{}: {}".format(i+1, X[i].value()))

print('Objective function value: f = {}'.format(lp4.objective.value()))
```

```python
lp6 = pulp.LpProblem('lp6', sense=pulp.LpMinimize)

p=[]

J = 10
M = 5

x=[]
st=[]
ct=[]
CT=pulp.LpVariable('CT', lowBound=0, cat="Continuous")

# 処理時間をランダムに決定&決定変数を登録．
for j in range(J):
  pt=[]
  st1=[]
  ct1=[]
  for m in range(M):
    pt.append(random.randint(1,10))
    st1.append(pulp.LpVariable('st{},{}'.format(j+1, m+1), lowBound=0, cat="Continuous"))
    ct1.append(pulp.LpVariable('ct{},{}'.format(j+1, m+1), lowBound=0, cat="Continuous"))
  p.append(pt)
  st.append(st1)
  ct.append(ct1)

# 決定変数を登録．
for i in range(J):
  xi=[]
  for j in range(J):
    xi.append(pulp.LpVariable('X{},{}'.format(i+1, j+1), cat="Binary"))
  x.append(xi)

################################

# 完了時刻の最小化
lp6 += ct[J-1][M-1] # Objective Function

# 開始時刻＋処理時間＝完了時刻
for i in range(J):
    for m in range(M):
        eq=None
        for j in range(J):
            eq+= p[j][m] * x[i][j]
        lp6+= st[i][m] + eq - ct[i][m] == 0

# 一度に1つの仕事しか処理できない
for i in range(J):
    eq=None
    for j in range(J):
        eq+= x[i][j]
    lp6+= eq == 1

# 1つの仕事は1回だけ処理される．
for j in range(J):
    eq=None
    for i in range(J):
        eq+= x[i][j]
    lp6+= eq == 1

# 開始時間はmax(前ジョブ，前工程)
for i in range(0,J):
    for m in range(0,M):
        if i>=1:
            lp6 += st[i][m] - ct[i-1][m] >=0
        if m>=1:
            lp6 += st[i][m] - ct[i][m-1] >=0

# print(lp6)

status = lp6.solve()

print('\n\nOptimal solution')
for i in range(J):
    for j in range(J):
        print("x{},{}: {}".format(i+1,j+1, x[i][j].value()))

# for i in range(J):
#     print("st{},{}: {}".format(i+1,M, st[i][M-1].value()))

# for i in range(J):
#     print("ct{},{}: {}".format(i+1,M, ct[i][M-1].value()))

print('Objective function value: f = {}'.format(lp6.objective.value()))
```
